```python
%pylab inline
import scipy.stats as stats
import numpy as np
```

```python
def function(sample):
    _x = mean(sample)
    return exp(_x)

def sample_with_replacement(my_array, n=None):

    sample = []
    if not n:
        n = len(my_array)

    for i in range(n):
        r = randint(len(my_array))
        sample.append(my_array[r])
    return np.array(sample)

def boot(sample_data, statistic_func, N=None, B=1000):
    N = N or len(sample_data)
    bootstrap_data = []
    for i in range(B):
        resampled_data = sample_with_replacement(sample_data)
        resampled_stat = statistic_func(resampled_data)
        bootstrap_data.append(resampled_stat)
    return bootstrap_data

def normal_interval(alpha, theta, boot_data):
    se = std(boot_data)
    z = stats.norm.ppf(1-alpha/2)
    return (theta - se*z, theta + se*z)
```

```python
data = np.random.normal(5,1,100)
```

```python
func = function
alpha = 0.05
theta = func(data)
boot_data = boot(data ,func)
boot_means = [sample_with_replacement(data).mean() for i in range(1000)]
ans = normal_interval(alpha,theta,boot_data)
```
