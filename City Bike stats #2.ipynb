```python
%pylab inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import sample
from scipy.stats import probplot

plt.rcParams['figure.figsize'] = 12,12
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.facecolor'] = 'f5f5f5'
```

###The Gender of riders follows the central limit theorem##
### The data must be sanitized if not the mean statistic will be usless##
###Once sanatized the mean statistic gives an accurate value to the ratio of men and women##

```python
df = pd.read_csv('201412-citibike-tripdata.csv')
df_clean1 = df[df.gender > 0]
```

```python
data1 = []
for i in xrange(1000):
    data1.append(df_clean1.loc[sample(df_clean1.index, 75)]['gender'].mean())
    
_ = hist(data1, bins=30, normed=True, alpha=0.4, label="Histogram")

mu = mean(data1)
s2 = var(data1)
X = linspace(min(data1), max(data1), 1000)
Y = 1.0 / sqrt(2*pi*s2) * e**(-(X-mu)**2 / (2*s2))
plot(X,Y,'r--')
```
