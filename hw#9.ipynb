```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
%pylab inline


np.random.seed(0)
df = pd.read_csv('cardata.csv')
x = array(df['HP'])
f = lambda t : 5*t**8+3*t**2+5*t**3+9*t**2+13

y = array(df['MPG'])
plt.scatter(x,y,s=4)


# first order fit
model = LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis],y)
x_test = np.linspace(-1,5,300)[:,np.newaxis]
y_test = model.predict(x_test)

plt.figure()
plt.scatter(x,y,s=4)
plt.plot(x_test,y_test,'r')
plt.axis('tight')
plt.legend(loc='best')
plt.show()


# In[14]:

# 6th order fit
x_test = np.linspace(-1,.2,300)
plt.figure()
plt.scatter(x,y,s=4)

X = np.array([x**i for i in range(7)]).T
X_test = np.array([x_test**i for i in range(7)]).T
order5 = LinearRegression()
order5.fit(X.squeeze(),y)
plt.plot(x_test,order5.predict(X_test),'r')
plt.axis('tight')
plt.show()
```
