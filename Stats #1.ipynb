```python
__author__ = 'NJackson'
__version__ = '1'


import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats as stats
from sklearn import linear_model
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.titlesize'] = 25
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['figure.edgecolor'] = 'c'
plt.rcParams['figure.facecolor'] = 'gray'
colors = plt.rcParams['axes.color_cycle']
%matplotlib inline
```

#Plotting Binomial Distribution pmf#
 ##$n$ vs pmf($n$)##

```python
plt.rcParams['figure.figsize'] = 14,8

k = range(3,1000)
for p, color in zip([0.3, 0.03], colors):
    rv = scipy.stats.binom(1000, p)
    plt.plot(k, rv.pmf(k), lw=6, color=color, label=p)
    plt.vlines(rv.mean(),0,.07,'r','--')
    plt.fill_between(k, rv.pmf(k), color=color, alpha=0.1)
plt.legend([0.3,0.03,'average'])
plt.title("$Binomial$ $distribution$")
plt.tight_layout()
plt.ylabel("PMF at $n$")
plt.xlabel("$n$")
plt.show()
```

#Scatter plot#
##number of heads vs $n$##

```python
ax1 = plt.subplot(221)
data = []
X_train = np.array(k[500:])[:,np.newaxis]
X = np.array(k[:500])[:,np.newaxis]
for i in k:
        data.append(np.random.binomial(i,.3))
Y_train = np.array(data[500:])
Y = np.array(data[:500])        
        

regr = linear_model.LinearRegression()        
regr.fit(X_train,Y_train)

plt.scatter(X,Y,c='b',s=20,alpha=.5)
plt.plot(X,regr.predict(X),'r--',linewidth=3)

ax2 = plt.subplot(222)
data = []
X_train = np.array(k[500:])[:,np.newaxis]
X = np.array(k[:500])[:,np.newaxis]
for i in k:
        data.append(np.random.binomial(i,.03))
Y_train = np.array(data[500:])
Y = np.array(data[:500])        
        

regr = linear_model.LinearRegression()        
regr.fit(X_train,Y_train)

plt.scatter(X,Y,c='b',s=20,alpha=.5)
plt.plot(X,regr.predict(X),'r--',linewidth=3)

ax1.set_title('0.3')
ax2.set_title('0.03')
```

#Q-Q plot#
##Behavior of distribution compared to a normal##

```python
ax1 = plt.subplot(221)
x = scipy.stats.binom.rvs(1000,.3,size=1000)
res =  stats.probplot(x, plot=plt)
ax1.set_title('0.3')

ax2 = plt.subplot(222)
x = scipy.stats.binom.rvs(1000,.03,size=1000)
res = scipy.stats.probplot(x, plot= plt)
ax2.set_title('0.03')
```
