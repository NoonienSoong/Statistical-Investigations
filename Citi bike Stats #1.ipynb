```python
def mean_plot(size):
    data = []
    data1 = []
    means = []
    means1 = []
    for i in range(size):
        x = np.random.normal(0,1)
        data.append(x)
        m = sum(data)/len(data)
        means.append(m)
    for i in range(size):
        x1 = np.random.standard_cauchy()
        data1.append(x1)
        m1 = sum(data1)/len(data1)
        means1.append(m1)
    plt.plot(means)
    plt.plot(means1)
    plt.ylabel('$Sample$ $Means$',size=40)
    plt.xlabel('$n$',size=50)
```

The Cauchy Distrobution does not have a converging mean. Shown in the graph below

```python
mean_plot(10000)
```

#** The Probability that a biker ends in Manhattan given they started in Brooklyn**#
###A = the biker ends in Manhattan###
###B = the biker starts in Brooklyn###
##$$P(A|B) = \frac{P(A,B)}{P(B)}$$##
<img src ='files/citibike-nyc-bike-share-docking-stations-map.jpg'>

#This problem was solved by:#

**1**. Choosing landmarks to divide the city up by (See picture) this allows me to seperate Bk and Manhattan

**2**. Find all bikes that were rented from Bk and brought to Manhattan. Because of the geography I had to 
   find all bikes that started to the right of the "NEast_River" longitude (i.e Williamsburg and parts of southern Bk)
   and see how many ended in manhattan that had to be done twice since the way the city was divided I first found the 
   bikes that ended in the northern part of manhattan and then the southern. These to numbers were then added togther
   for the southern part of Bk (or Downtown Bk) the same process was done
 
**3**. Then I found my reduced sample space which in this case was just all bikes that started in Bk

**4**. At last dividing the finding I came to my result


```python
NEast_river =geo.Geocoder.geocode('East River Park')
Bk_bridge = geo.Geocoder.geocode('Empire Fulton Ferry')
Rd_hook = geo.Geocoder.geocode('Brooklyn Bridge Park Pier 5')
```

```python
#lat     |    long
Bk_bridge.coordinates
# (40.704224, -73.9933767)
NEast_river.coordinates
# (40.7153271, -73.9754516)
Rd_hook.coordinates
# (40.6948689, -74.0017282)
```

Everything that starts right of North East river latitutude (i.e. williamsburg and 
parts of southern BK) and ends in manhattan

```python
North_Bk1 = len(df[df['start station longitude'] > -73.972448][df['start station latitude'] < 40.719995][df['end station longitude'] < -73.972448][df['end station latitude'] > 40.703543])
North_bk2 = len(df[df['start station longitude'] > -73.972448][df['start station latitude'] < 40.719995][df['end station longitude'] < -74.0017282][df['end station latitude'] < 40.703543]) 
North_bk = North_Bk1 + North_bk2
```

Everthing that starts in downtown bk and ends in manhattan

```python
South_bk1 = len(df[df['start station longitude'] < -73.972448][df['start station longitude'] > -74.0017282][df['start station latitude'] < 40.704224][df['end station longitude'] < -73.972448 ][df['end station latitude'] > 40.703543])
South_bk2 = len(df[df['start station longitude'] < -73.972448][df['start station longitude'] > -74.0017282][df['start station latitude'] < 40.704224][df['end station longitude'] < -74.0017282 ][df['end station latitude'] < 40.703543])
South_bk = South_bk1 + South_bk2
```

now finding bikes that start in bk

```python
Start_Nbk = len(df[df['start station longitude'] > -73.972448][df['start station latitude']< 40.719995]) 
Start_Sbk = len(df[df['start station longitude'] < -73.972448][df['start station longitude'] > -74.0017282][df['start station latitude'] < 40.703543])
Start_bk = Start_Nbk + Start_Sbk
```

###The answer:###

```python
P = (North_bk+South_bk)/Decimal(Start_bk)
```
